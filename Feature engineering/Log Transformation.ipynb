```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.array([1, 10, 100, 1000])
```

```python
sns.displot(data,kind='kde')
```

```python
log_data = np.log(data)  # Natural logarithm (base e)
log_data 
```

```python
sns.displot(log_data,kind='kde')
```

```python
# Base 10 logarithm
log_data_base10 = np.log10(data)
log_data_base10
```

```python
sns.displot(log_data_base10,kind='kde')
```

```python
print("Original data:", data)
print("Natural logarithm:", log_data)
print("Base 10 logarithm:", log_data_base10)
```

```python
log_base_2 =  np.log2(data)
log_base_2
```

```python
sns.distplot(log_base_2)
```

```python
# Generate 1000 samples from a right-skewed distribution
data = np.random.exponential(scale=2, size=1000)
data
```

```python
plt.hist(data, bins=50)
plt.show()
```

```python
# Apply log transformation to the data
log_data = np.log(data)
```

```python
plt.hist(data, bins=50, label='Original Data')
plt.legend()
```

```python
plt.hist(log_data, bins=50, label='Log-Transformed Data')
plt.legend()
plt.show()
```

```python
# Apply log transformation to the data
log_data_2 = np.log2(data)
log_data_2
```

```python
plt.hist(log_data_2, bins=50, label='Log 2-Transformed Data')
plt.legend()
plt.show()
```
