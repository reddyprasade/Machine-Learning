# Normalization (or min-max normalization)

```python
import numpy as np
```

```python
def min_max_normalization(data):
    min_val = np.min(data)
    max_val = np.max(data)

    x = data.copy()
    # Perform Min-Max scaling
    scaled_data = (x - min_val)/(max_val-min_val)
    return scaled_data
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset/healthcare-dataset-stroke-data.csv")
df.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df,kind='kde')
```

```python
bmi  = df['bmi']
bmi
```

```python
sns.displot(bmi,kind='kde')
```

```python
bmi_scal = min_max_normalization(bmi)
bmi_scal
```

```python
sns.displot(bmi_scal,kind='kde')
```

```python
data = np.array([1, 5, 10, 15, 20,-20])
```

```python
norm = min_max_normalization(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 8))

# Plot original data
plt.subplot(1, 2, 1)
plt.plot(data, 'bo-', label='Original Data')
plt.title('Original Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Plot normalized data
plt.subplot(1, 2, 2)
plt.plot(norm, 'ro-', label='Normalized Data')
plt.title('Normalized Data')
plt.xlabel('Index')
plt.ylabel('Value (Normalized)')
plt.legend()

plt.tight_layout()
plt.show()
```
